```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

# In this part is the data

```python
años = np.array([2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022], dtype=float)
Iphone = np.array([849000, 889000, 900000, 909000, 120000, 1319000, 1190000, 1400000], dtype=float)
Samsun = np.array([678568, 697983, 785352, 824183, 882429, 892137, 1222199, 1319276], dtype=float)
```

##### print("truth values of second slice:" , años)
print("truth values of second slice:" , Iphone)
print("truth values of second slice:" , Samsun)

# Here is the model of the dataset

```python
oculta1 = tf.keras.layers.Dense(units=3, input_shape=[1])
oculta2 = tf.keras.layers.Dense(units=3)
oculta3 = tf.keras.layers.Dense(units=3)
salida = tf.keras.layers.Dense(units=1)

modelo = tf.keras.Sequential([oculta1, oculta2, oculta3, salida])
```

```python
if np.array([]).size: print(1)
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
    
```

# here is the training of the model

```python
print("Comenzando entrenamiento...")
historial = modelo.fit(Samsun, Iphone, epochs=1000, batch_size=32)
print("Modelo entrenado!")
```

# Here es the graphic of error

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(historial.history["loss"])
```

```python
print("Hagamos una predicción!")
resultado = modelo.predict([1319276])
print("El resultado es " + str(resultado) )
```
